```python
#Decision Tree
import numpy as np
from sklearn import tree
from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier, plot_tree
import matplotlib.pyplot as plt
import math
import pandas as pd
```

```python
#Load csv and store as DataFrames
df_white = pd.read_csv('winequality-red.csv')
df_red = pd.read_csv('winequality-red.csv')
```

```python
#df_white.head()
```

```python
# Split Whiite Wine into training and testing datasets

row_count = df_white.shape[0]
train = round(.7*row_count)
w_train_features = df_white.iloc[:train,:-1]
w_test_features = df_white.iloc[train:,:-1]
w_train_targets = df_white.iloc[:train,-1] #-1 represents last column which is quality (target)
w_test_targets = df_white.iloc[train:,-1]
```

```python
#Split Red Wine into training and testing datasets

r_row_count = df_red.shape[0]
r_train = round(.7*r_row_count)
r_train_features = df_red.iloc[:r_train,:-1]
r_test_features = df_red.iloc[r_train:,:-1]
r_train_targets = df_red.iloc[:r_train,-1] #-1 represents last column which is quality (target)
r_test_targets = df_red.iloc[r_train:,-1]
```

```python
#White wine tree
w_tree = DecisionTreeClassifier(criterion = 'entropy').fit(w_train_features,w_train_targets)
```

```python
#white wine prediction
prediction = w_tree.predict(w_test_features)
```

```python
#white wine accuracy
print("The prediction accuracy for white wine is: ",w_tree.score(w_test_features,w_test_targets)*100,"%")
```

```python
#Plot White Wine decision tree
X = w_test_features
y = w_test_targets
plt.figure(figsize=(10, 10))
clf = DecisionTreeClassifier().fit(X, y)
dtree = plot_tree(clf, filled = True)
plt.show()
```

```python
#red wine tree
r_tree = DecisionTreeClassifier(criterion = 'entropy').fit(r_train_features,r_train_targets)
```

```python
#red wine prediction
prediction = r_tree.predict(r_test_features)
```

```python
#red wine accuracy
print("The prediction accuracy for white wine is: ",r_tree.score(r_test_features,r_test_targets)*100,"%")
```

```python
#Plot Red Wine decision tree
X_r = r_test_features
y_r = r_test_targets
plt.figure(figsize=(10, 10))
red_clf = DecisionTreeClassifier().fit(X_r, y_r)
dtree = plot_tree(red_clf, filled = True)
plt.show()
```
